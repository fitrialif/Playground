This notebook explores how the Principal Component Analysis works by making use of synthetic data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
def random_df(mx=0, sx=1, my=0, sy=1, n=1000):
    '''Generates a random 2D normal dataframe'''
    x = np.random.normal(mx, sx, 1000)
    y = np.random.normal(mx, sy, 1000)
    
    return pd.DataFrame({'x': x, 'y':y})

def plot_xy(x,y):
    '''Makes a scatter plot'''
    plt.scatter(x, y)
    M = max(x.max(), y.max())
    m = min(x.min(), y.min())
    plt.xlim(m, M)
    plt.ylim(m, M)
    #plt.gca().set_aspect('equal', adjustable='box')
    plt.show()
    
def pca_components(df=None, mx=0, sx=1, my=0, sy=1, n=1000, x='x', y='y'):
    '''Generates a dataset if needed, performs PCA and prints the components and variance ratio'''
    if df is None:
        df = random_df(mx, sx, my, sy, n)
    plot_xy(df[x], df[y])
    pca = PCA()
    pca.fit(df)
    print('Components:')
    print(pca.components_)
    print('Variance ratio:')
    print(pca.explained_variance_ratio_)
```

# Two normal uncorrelated variables
The variance ratio of the components is almost the same, meaning that one needs to have information for each variable to define the dataset. The components are composed by a biased weight distribution.

```python
pca_components()
```

# One normal variable and one constant
All the variance is explained by the normally distributed variable.

```python
pca_components(mx=0, sx=10, my=0, sy=0)
```

# Two normal uncorrelated variables with different variances
The variance ratio of the components leans to the higher variance distribution.

```python
pca_components(mx=0, sx=10, my=0, sy=1)
```

```python
pca_components(mx=0, sx=1, my=0, sy=10)
```

# One normal and one chi squared distribution
The variance ratio of the components is almost the same, meaning that one needs to have information for each variable to define the dataset. The components are composed by a very biased weight distribution, meaning that each variable has a very different behaviour than the other.

```python
pca_components(pd.DataFrame({'x': np.random.chisquare(1,1000), 'y':np.random.normal(0, 1, 1000)}))
```

# Two correlated chisquared variables
The variance ratio of the components leans highly to one of the variables, meaning that if we have information for that variable then we know how the dataset varies entirely. Nevertheless, the components of the first PC have about an equal contribution of each variable, meaning that we need to have the levels of both to get all the information in.

```python
x = np.random.chisquare(50,1000)
pca_components(pd.DataFrame({'x': x, 'y': x * 1.12}))
```
